```python
import torch 
import numpy as np 
```

```python
x_train = torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
y_train = torch.FloatTensor([[0],[1],[1],[0]])
```

```python
W_h = torch.randn([2,30], requires_grad=True)
b_h = torch.randn([30], requires_grad=True)
W_o = torch.randn([30,1], requires_grad=True)
b_o = torch.randn([1], requires_grad=True)
```

```python
optimizer = torch.optim.SGD([W_h, b_h, W_o, b_o],lr=0.01)
```

```python
def model_ANN(x):
  HLl = torch.sigmoid(torch.matmul(x, W_h) + b_h)
  Out = torch.sigmoid(torch.matmul(HLl, W_o) + b_o)
  return Out 
```

```python
for step in range(200000):
  prediction = model_ANN (x_train)
  cost = torch.mean((-1) * y_train * torch.log(prediction) + (-1) * (1 - y_train) * torch.log(1 - prediction))
  if step%10000 == 0:
    print("step", step,"cost", cost.detach().numpy())
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
```

```python
model_test = model_ANN(x_train)
print(model_test.detach().numpy())
```
